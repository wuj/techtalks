# Attention Data Preparation Transformer Explorer Demo (Script 2)
Runs demo sentences through **BERT-base-uncased** and extracts everything
the Session 2 HTML/JS demo needs:

| Data | Purpose in JS demo |
|------|--------------------|
| Attention weights (all layers/heads) | Heatmap panel: click a token, see who it attends to |
| Raw Q·K scores (all layers/heads) | Hover tooltip + softmax temperature slider |
| Q / K / V vectors (selected layers) | Colored bar charts alongside each token |
| Contextual nearest neighbors | Before/after panel: how "bank" moves in meaning |
| 2D sentence projections (per layer) | Animate token positions shifting across layers |

**Produces:** `attention_data.json` (~3-5 MB)

**Expected runtime:** ~2 min first run (model download), ~30 s thereafter.

**Why BERT?** BERT is bidirectional: every token can attend to every
other token. This is essential for the "bank" demo: bank needs to
look *forward* at "river."

````python
# ============================================================
# 2. CONFIGURATION
# ============================================================
import numpy as np
import json
import math
from pathlib import Path

# --- Model ---
MODEL_NAME = "bert-base-uncased"

# --- Output ---
OUTPUT_DIR = Path("/home/jovyan/work/playground/notebooks/tech_talks/data")
OUTPUT_DIR.mkdir(exist_ok=True)

# --- Which layers to export full Q/K/V vectors for ---
# (Attention weights and raw QK scores are exported for ALL layers.)
# "auto" = [0, best_layer, last_layer], determined after best-layer detection.
QKV_EXPORT_LAYERS = "auto"

# --- Nearest neighbors ---
N_NEIGHBORS = 20          # per token of interest, per layer
MIN_REF_WORD_LEN = 2      # skip single-letter tokens in reference vocab

# --- Precision (decimal places) ---
ATTN_PRECISION = 4         # attention weights (probabilities 0–1, need precision near 0)
SCORE_PRECISION = 2        # raw QK scores (logits, coarser is fine for tooltips)
QKV_PRECISION = 4          # Q/K/V vector components
NEIGHBOR_PRECISION = 4     # cosine similarities
PROJ_PRECISION = 2         # 2D projection coordinates

RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)
````

````python
# ============================================================
# 3. LOAD BERT MODEL + TOKENIZER
# ============================================================
import torch
from transformers import BertModel, BertTokenizer

print(f"Loading '{MODEL_NAME}' ...")
print("(First run downloads ~440 MB. Subsequent runs use cache.)\n")

tokenizer = BertTokenizer.from_pretrained(MODEL_NAME)
model = BertModel.from_pretrained(
    MODEL_NAME,
    output_attentions=True,
    output_hidden_states=True,
)
model.eval()

n_layers = model.config.num_hidden_layers
n_heads = model.config.num_attention_heads
hidden_dim = model.config.hidden_size
head_dim = hidden_dim // n_heads

print(f"✓ {MODEL_NAME}")
print(f"  Layers: {n_layers}, Heads: {n_heads}, "
      f"Hidden: {hidden_dim}, Head dim: {head_dim}")
print(f"  Vocab size: {tokenizer.vocab_size:,}")
````

````python
# ============================================================
# 4. DEFINE DEMO SENTENCES
# ============================================================
# Each sentence specifies:
#   - tokens_of_interest: which tokens to track (with occurrence index
#     for duplicates like the two "bank" tokens)
#   - scoring_hints: token texts used to auto-score which attention head
#     best discriminates river vs. finance context

SENTENCES = {
    "bank_combined": {
        "text": "The bank by the river was steep, but the bank approved her loan quickly.",
        "tokens_of_interest": [
            {
                "token": "bank",
                "occurrence": 1,
                "label": "bank₁ (river)",
                "context_type": "river",
            },
            {
                "token": "bank",
                "occurrence": 2,
                "label": "bank₂ (finance)",
                "context_type": "finance",
            },
        ],
        "scoring_hints": {
            "river_tokens": ["river", "steep", "by"],
            "finance_tokens": ["approved", "loan", "quickly"],
        },
    },
    "bank_river": {
        "text": "I sat on the bank of the river.",
        "tokens_of_interest": [
            {
                "token": "bank",
                "occurrence": 1,
                "label": "bank (river)",
                "context_type": "river",
            },
        ],
        "scoring_hints": {
            "river_tokens": ["sat", "river"],
            "finance_tokens": [],
        },
    },
    "bank_finance": {
        "text": "I deposited cash at the bank.",
        "tokens_of_interest": [
            {
                "token": "bank",
                "occurrence": 1,
                "label": "bank (finance)",
                "context_type": "finance",
            },
        ],
        "scoring_hints": {
            "river_tokens": [],
            "finance_tokens": ["deposited", "cash"],
        },
    },
}

print(f"✓ {len(SENTENCES)} demo sentences configured")
````

````python
# ============================================================
# 5. TOKENIZE + VALIDATE
# ============================================================
# Tokenize each sentence, find the positions of tokens of interest,
# and resolve scoring-hint words to token indices.

for sid, sdata in SENTENCES.items():
    encoded = tokenizer(sdata["text"], return_tensors="pt")
    token_ids = encoded["input_ids"][0].tolist()
    tokens = tokenizer.convert_ids_to_tokens(token_ids)

    sdata["encoded"] = encoded
    sdata["token_ids"] = token_ids
    sdata["tokens"] = tokens
    sdata["special_token_mask"] = [
        tok in ("[CLS]", "[SEP]", "[PAD]") for tok in tokens
    ]

    print(f'\n--- {sid} ({len(tokens)} tokens) ---')
    print(f'  Text:   "{sdata["text"]}"')
    print(f"  Tokens: {tokens}")

    # --- Locate each token of interest ---
    for toi in sdata["tokens_of_interest"]:
        target = toi["token"].lower()
        occ = toi["occurrence"]
        count = 0
        found = False
        for idx, tok in enumerate(tokens):
            if tok == target:
                count += 1
                if count == occ:
                    toi["index"] = idx
                    found = True
                    break
        if not found:
            # Check if the token was split into subwords
            print(f'  ⚠ "{toi["token"]}" (occurrence {occ}) NOT found as a single token!')
            print(f"    Tokens are: {tokens}")
            print(f"    The word may have been split by WordPiece. Check manually.")
        else:
            print(f'  ✓ "{toi["label"]}" → index {toi["index"]}  (token: "{tokens[toi["index"]]}")')

    # --- Resolve scoring-hint words to token indices ---
    resolved_hints = {}
    for group_name, hint_words in sdata["scoring_hints"].items():
        indices = []
        for hw in hint_words:
            hw_lower = hw.lower()
            for idx, tok in enumerate(tokens):
                if tok == hw_lower:
                    indices.append(idx)
                    break
        resolved_hints[group_name] = indices
    sdata["scoring_indices"] = resolved_hints
    print(f"  Scoring indices: { {k: list(zip(sdata['scoring_hints'][k], v)) for k, v in resolved_hints.items()} }")
````

````python
# ============================================================
# 6. FORWARD PASS — extract attention weights + hidden states
# ============================================================
# BERT architecture recap:
#   hidden_states[0]  = embedding output (word + position + segment + LayerNorm)
#   hidden_states[i]  = output of encoder layer (i-1)  (for i = 1..12)
#   attentions[i]     = attention weights from encoder layer i  (for i = 0..11)
#                       shape: [batch, n_heads, seq_len, seq_len]

print("Running forward passes ...\n")

for sid, sdata in SENTENCES.items():
    with torch.no_grad():
        outputs = model(**sdata["encoded"])

    # hidden_states: tuple of 13 tensors, each [1, seq_len, 768]
    # Convert to list of numpy arrays [seq_len, 768] (drop batch dim)
    sdata["hidden_states"] = [
        h[0].numpy().copy() for h in outputs.hidden_states
    ]

    # attentions: tuple of 12 tensors, each [1, n_heads, seq_len, seq_len]
    # Convert to list of numpy arrays [n_heads, seq_len, seq_len]
    sdata["attentions"] = [
        a[0].numpy().copy() for a in outputs.attentions
    ]

    seq_len = len(sdata["tokens"])
    print(f"  ✓ {sid}: {seq_len} tokens, "
          f"hidden_states[0].shape={sdata['hidden_states'][0].shape}, "
          f"attentions[0].shape={sdata['attentions'][0].shape}")
````

````python
# ============================================================
# 7. COMPUTE Q / K / V + RAW QK SCORES
# ============================================================
# For each layer, manually compute Q, K, V by applying the layer's
# linear projections to the input hidden state, then compute the
# scaled dot-product scores (before softmax).
#
# We verify correctness by checking that softmax(scores) ≈ attention weights.

print("Computing Q/K/V and raw QK scores ...\n")

for sid, sdata in SENTENCES.items():
    sdata["qkv_all"] = {}        # layer_idx → {q, k, v} each [n_heads, seq_len, head_dim]
    sdata["qk_scores_all"] = {}  # layer_idx → [n_heads, seq_len, seq_len]

    for layer_idx in range(n_layers):
        layer = model.encoder.layer[layer_idx]
        attn_module = layer.attention.self

        # Input to this layer's self-attention (explicit float32)
        hidden = torch.as_tensor(
            sdata["hidden_states"][layer_idx], dtype=torch.float32
        ).unsqueeze(0)

        with torch.no_grad():
            # Linear projections (includes bias)
            Q = attn_module.query(hidden)[0]   # [seq_len, hidden_dim]
            K = attn_module.key(hidden)[0]
            V = attn_module.value(hidden)[0]

            seq_len = Q.shape[0]

            # Reshape to per-head: [seq_len, hidden] → [n_heads, seq_len, head_dim]
            Q = Q.view(seq_len, n_heads, head_dim).permute(1, 0, 2)
            K = K.view(seq_len, n_heads, head_dim).permute(1, 0, 2)
            V = V.view(seq_len, n_heads, head_dim).permute(1, 0, 2)

            # Scaled dot-product scores (before softmax)
            scores = torch.matmul(Q, K.transpose(-1, -2)) / math.sqrt(head_dim)
            # scores shape: [n_heads, seq_len, seq_len]

        sdata["qkv_all"][layer_idx] = {
            "q": Q.numpy().copy(),
            "k": K.numpy().copy(),
            "v": V.numpy().copy(),
        }
        sdata["qk_scores_all"][layer_idx] = scores.numpy().copy()

    # --- Verify: softmax(scores) should match the model's attention weights ---
    # Check one layer as a spot-check
    check_layer = 0
    recomputed = torch.softmax(
        torch.tensor(sdata["qk_scores_all"][check_layer]), dim=-1
    ).numpy()
    model_attn = sdata["attentions"][check_layer]
    max_diff = np.abs(recomputed - model_attn).max()
    print(f"  {sid}: QK verification (layer {check_layer}) max|diff| = {max_diff:.2e} ", end="")
    print("✓" if max_diff < 1e-5 else "⚠ MISMATCH — check attention mask or scaling")

print("\n✓ Q/K/V computation complete.")
````

````python
# ============================================================
# 8. BUILD REFERENCE VOCABULARY FROM BERT'S EMBEDDING TABLE
# ============================================================
# For nearest-neighbor search, we compare each token's contextual
# hidden state against BERT's own input word embeddings. This asks:
# "After contextual processing, which vocabulary word's static
# embedding is this vector closest to?"
#
# We filter BERT's vocabulary to clean, whole-word tokens only.

word_embeddings = model.embeddings.word_embeddings.weight.detach().numpy()
# shape: [vocab_size, hidden_dim]

# Build filtered reference vocabulary
ref_vocab = []        # list of (word, index_in_bert_vocab)
ref_embeddings = []   # corresponding embedding vectors

for token_str, token_id in tokenizer.vocab.items():
    # Skip subword fragments (## prefix)
    if token_str.startswith("##"):
        continue
    # Skip special tokens
    if token_str in ("[CLS]", "[SEP]", "[PAD]", "[MASK]", "[UNK]"):
        continue
    # Only alphabetic, minimum length
    if not token_str.isalpha():
        continue
    if len(token_str) < MIN_REF_WORD_LEN:
        continue

    ref_vocab.append((token_str, token_id))
    ref_embeddings.append(word_embeddings[token_id])

ref_embeddings = np.array(ref_embeddings, dtype=np.float32)

# Pre-compute norms for fast cosine similarity
ref_norms = np.linalg.norm(ref_embeddings, axis=1, keepdims=True)
ref_norms = np.where(ref_norms < 1e-10, 1.0, ref_norms)  # avoid div-by-zero
ref_embeddings_normed = ref_embeddings / ref_norms

print(f"✓ Reference vocabulary: {len(ref_vocab):,} clean words "
      f"(from {tokenizer.vocab_size:,} total BERT tokens)")
print(f"  Embedding shape: {ref_embeddings.shape}")

# Quick sanity: check that "bank", "river", "loan" are in ref vocab
for probe in ["bank", "river", "loan", "deposit", "shore", "credit"]:
    found = any(w == probe for w, _ in ref_vocab)
    print(f"  {'✓' if found else '⚠'} '{probe}' in reference vocab: {found}")
````

````python
# ============================================================
# 9. COMPUTE NEAREST NEIGHBORS (static + contextual per layer)
# ============================================================
# For each token of interest:
#   - Static neighbors: BERT input embedding vs reference vocab
#   - Contextual neighbors at each layer: hidden state vs reference vocab

def cosine_neighbors(query_vec, topn=N_NEIGHBORS, exclude_words=None):
    """Find nearest neighbors in the reference vocabulary by cosine similarity."""
    exclude_words = set(exclude_words or [])
    q = query_vec / (np.linalg.norm(query_vec) + 1e-10)
    sims = ref_embeddings_normed @ q   # [n_ref]

    # Get top indices
    top_k = min(topn + len(exclude_words) + 5, len(ref_vocab))  # grab extras for filtering
    top_indices = np.argpartition(-sims, top_k)[:top_k]
    top_indices = top_indices[np.argsort(-sims[top_indices])]

    results = []
    for idx in top_indices:
        word = ref_vocab[idx][0]
        if word in exclude_words:
            continue
        results.append({"word": word, "similarity": round(float(sims[idx]), NEIGHBOR_PRECISION)})
        if len(results) >= topn:
            break
    return results


print("Computing nearest neighbors ...\n")

for sid, sdata in SENTENCES.items():
    for toi in sdata["tokens_of_interest"]:
        tok_idx = toi["index"]
        tok_text = toi["token"].lower()

        # --- Static neighbors (from BERT's input embedding table) ---
        # Look up the token's word embedding (no position or segment)
        bert_token_id = sdata["token_ids"][tok_idx]
        static_vec = word_embeddings[bert_token_id]
        toi["static_neighbors"] = cosine_neighbors(static_vec, exclude_words={tok_text})

        print(f'  {sid} / {toi["label"]} — static neighbors:')
        for nb in toi["static_neighbors"][:5]:
            print(f'    {nb["word"]:15s} {nb["similarity"]:.4f}')

        # --- Contextual neighbors at each layer ---
        toi["contextual_neighbors"] = {}
        for layer_idx in range(n_layers + 1):
            # hidden_states[0] = embedding output, [1]..[12] = encoder layers
            ctx_vec = sdata["hidden_states"][layer_idx][tok_idx]
            toi["contextual_neighbors"][layer_idx] = cosine_neighbors(
                ctx_vec, exclude_words={tok_text}
            )

        # Show a few key layers
        for show_layer in [0, n_layers // 2, n_layers]:
            top3 = [nb["word"] for nb in toi["contextual_neighbors"][show_layer][:3]]
            print(f'    Layer {show_layer:2d}: {", ".join(top3)} ...')
        print()
````

````python
# ============================================================
# 10. AUTO-SELECT BEST LAYER AND HEAD
# ============================================================
# For the combined sentence, score each (layer, head) by how well
# it discriminates the two "bank" meanings:
#
#   score = Σ_toi  ( attn[toi → expected_context] − attn[toi → wrong_context] )
#
# The head with the highest score best separates bank₁ (river) from bank₂ (finance).

print("Auto-detecting best layer/head for bank disambiguation ...\n")

best_defaults = {}

for sid, sdata in SENTENCES.items():
    river_indices = sdata["scoring_indices"].get("river_tokens", [])
    finance_indices = sdata["scoring_indices"].get("finance_tokens", [])

    # Skip scoring if we don't have both context types
    if not river_indices or not finance_indices:
        # For single-sense sentences, pick the head where "bank" attends most
        # to the matching context tokens
        toi = sdata["tokens_of_interest"][0]
        expected_type = toi["context_type"]
        expected_indices = sdata["scoring_indices"].get(f"{expected_type}_tokens", [])
        if expected_indices:
            best_score = -1
            best_lh = (0, 0)
            for li in range(n_layers):
                attn = sdata["attentions"][li]  # [n_heads, seq, seq]
                for hi in range(n_heads):
                    score = sum(attn[hi, toi["index"], ei] for ei in expected_indices)
                    if score > best_score:
                        best_score = score
                        best_lh = (li, hi)
            best_defaults[sid] = {"layer": best_lh[0], "head": best_lh[1]}
            print(f"  {sid}: best = layer {best_lh[0]}, head {best_lh[1]} "
                  f"(score={best_score:.3f}, single-sense heuristic)")
        else:
            best_defaults[sid] = {"layer": n_layers // 2, "head": 0}
            print(f"  {sid}: no scoring hints, defaulting to layer {n_layers // 2}, head 0")
        continue

    # --- Combined sentence: full discrimination scoring ---
    scores_grid = np.zeros((n_layers, n_heads))

    for li in range(n_layers):
        attn = sdata["attentions"][li]  # [n_heads, seq, seq]
        for hi in range(n_heads):
            score = 0.0
            for toi in sdata["tokens_of_interest"]:
                ti = toi["index"]
                attn_river = sum(attn[hi, ti, ri] for ri in river_indices)
                attn_finance = sum(attn[hi, ti, fi] for fi in finance_indices)
                if toi["context_type"] == "river":
                    score += attn_river - attn_finance
                else:
                    score += attn_finance - attn_river
            scores_grid[li, hi] = score

    best_li, best_hi = np.unravel_index(scores_grid.argmax(), scores_grid.shape)
    best_defaults[sid] = {"layer": int(best_li), "head": int(best_hi)}

    print(f"  {sid}: best = layer {best_li}, head {best_hi} "
          f"(discrimination score = {scores_grid[best_li, best_hi]:.3f})")

    # Also store per-token best for reference
    for toi in sdata["tokens_of_interest"]:
        toi["best_layer"] = int(best_li)
        toi["best_head"] = int(best_hi)

    # Show top-5 heads
    flat = scores_grid.flatten()
    top5 = np.argsort(-flat)[:5]
    print(f"  Top 5 heads:")
    for rank, fi in enumerate(top5):
        li, hi = np.unravel_index(fi, scores_grid.shape)
        print(f"    #{rank+1}: layer {li}, head {hi}  (score={scores_grid[li, hi]:.3f})")
    print()

# --- Resolve QKV export layers ---
if QKV_EXPORT_LAYERS == "auto":
    # Use the combined sentence's best layer, plus first and last
    best_layer = best_defaults.get("bank_combined", {}).get("layer", n_layers // 2)
    qkv_layers = sorted(set([0, best_layer, n_layers - 1]))
    print(f"Auto-selected QKV export layers: {qkv_layers}")
else:
    qkv_layers = sorted(set(QKV_EXPORT_LAYERS))
    print(f"QKV export layers (manual): {qkv_layers}")
````

````python
# ============================================================
# 11. COMPUTE 2D SENTENCE PROJECTIONS (PCA)
# ============================================================
# For each sentence, project all tokens' hidden states to 2D using
# a single PCA fit across all layers (so positions are comparable
# and we can animate movement across layers).

from sklearn.decomposition import PCA

print("Computing 2D sentence projections (PCA) ...\n")

for sid, sdata in SENTENCES.items():
    # Concatenate hidden states from all layers: [n_layers+1 × seq_len, hidden_dim]
    all_hidden = np.concatenate(sdata["hidden_states"], axis=0)

    pca = PCA(n_components=2, random_state=RANDOM_SEED)
    pca.fit(all_hidden)

    sdata["projections"] = {}
    for layer_idx in range(n_layers + 1):
        projected = pca.transform(sdata["hidden_states"][layer_idx])
        sdata["projections"][layer_idx] = projected

    print(f"  ✓ {sid}: PCA fit on {all_hidden.shape[0]} vectors, "
          f"explained variance = {pca.explained_variance_ratio_.sum():.2%}")
````

````python
# ============================================================
# 12. VALIDATION + VISUALIZATION
# ============================================================
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# We'll visualize the combined sentence's best head.
sid = "bank_combined"
sdata = SENTENCES[sid]
best = best_defaults[sid]
bl, bh = best["layer"], best["head"]
tokens = sdata["tokens"]
seq_len = len(tokens)

print(f"Visualizing: {sid}")
print(f"  Best layer={bl}, head={bh}\n")

fig = plt.figure(figsize=(18, 14))
gs = gridspec.GridSpec(2, 2, hspace=0.35, wspace=0.3)

# --- Panel 1: Attention heatmap (best head) ---
ax1 = fig.add_subplot(gs[0, 0])
attn = sdata["attentions"][bl][bh]  # [seq_len, seq_len]
im = ax1.imshow(attn, cmap="Blues", aspect="auto", vmin=0, vmax=attn.max())
ax1.set_xticks(range(seq_len))
ax1.set_xticklabels(tokens, rotation=45, ha="right", fontsize=7)
ax1.set_yticks(range(seq_len))
ax1.set_yticklabels(tokens, fontsize=7)
ax1.set_title(f"Attention weights — layer {bl}, head {bh}", fontsize=10)
ax1.set_xlabel("Key (attended to)")
ax1.set_ylabel("Query (attending from)")
plt.colorbar(im, ax=ax1, fraction=0.046)

# Highlight the two "bank" tokens
for toi in sdata["tokens_of_interest"]:
    ti = toi["index"]
    ax1.axhline(y=ti, color="red", linewidth=0.5, alpha=0.5)
    ax1.text(-0.5, ti, "◀", color="red", fontsize=8, ha="right", va="center")

# --- Panel 2: Attention arcs for both "bank" tokens ---
ax2 = fig.add_subplot(gs[0, 1])
ax2.set_xlim(-0.5, seq_len - 0.5)
ax2.set_ylim(-0.3, 1.3)

# Draw token labels along the bottom
for i, tok in enumerate(tokens):
    color = "red" if tok == "bank" else "black"
    weight = "bold" if tok == "bank" else "normal"
    ax2.text(i, 0, tok, ha="center", va="top", fontsize=7, color=color, fontweight=weight,
             rotation=45)

colors = ["#E74C3C", "#2980B9"]  # red for bank₁, blue for bank₂
for ci, toi in enumerate(sdata["tokens_of_interest"]):
    ti = toi["index"]
    attn_row = sdata["attentions"][bl][bh][ti]  # [seq_len]
    # Draw arcs for top-5 attended tokens
    top5 = np.argsort(-attn_row)[:6]
    for target in top5:
        if target == ti:
            continue
        weight = attn_row[target]
        if weight < 0.02:
            continue
        # Arc from ti to target
        mid = (ti + target) / 2
        height = abs(target - ti) / seq_len * 0.8 + 0.15
        y_offset = 0.15 + ci * 0.02
        arc_x = np.linspace(ti, target, 50)
        arc_y = y_offset + height * np.sin(np.linspace(0, np.pi, 50))
        ax2.plot(arc_x, arc_y, color=colors[ci], alpha=min(weight * 3, 0.9),
                 linewidth=weight * 8, solid_capstyle="round")
        ax2.text(target, y_offset + height + 0.02, f"{weight:.2f}",
                 ha="center", fontsize=5, color=colors[ci], alpha=0.7)

ax2.set_title(f"Attention arcs — layer {bl}, head {bh}", fontsize=10)
ax2.legend(
    [plt.Line2D([0], [0], color=c, lw=2) for c in colors],
    [toi["label"] for toi in sdata["tokens_of_interest"]],
    fontsize=8, loc="upper right",
)
ax2.axis("off")

# --- Panel 3: Nearest neighbors before/after ---
ax3 = fig.add_subplot(gs[1, 0])
ax3.axis("off")
ax3.set_title("Nearest neighbors — before vs. after (best layer)", fontsize=10)

y = 0.95
for toi in sdata["tokens_of_interest"]:
    ax3.text(0.0, y, toi["label"], fontsize=9, fontweight="bold",
             transform=ax3.transAxes)
    y -= 0.06

    # Static neighbors (before)
    static_top5 = [f'{nb["word"]} ({nb["similarity"]:.3f})'
                   for nb in toi["static_neighbors"][:5]]
    ax3.text(0.02, y, f"Static:      {', '.join(static_top5)}",
             fontsize=7, transform=ax3.transAxes, family="monospace")
    y -= 0.05

    # Contextual neighbors at best layer (after)
    ctx_layer = toi.get("best_layer", bl)
    ctx_top5 = [f'{nb["word"]} ({nb["similarity"]:.3f})'
                for nb in toi["contextual_neighbors"][ctx_layer][:5]]
    ax3.text(0.02, y, f"Layer {ctx_layer:2d}:    {', '.join(ctx_top5)}",
             fontsize=7, transform=ax3.transAxes, family="monospace",
             color="#E74C3C" if toi["context_type"] == "river" else "#2980B9")
    y -= 0.08

# --- Panel 4: 2D token positions at layer 0 vs. best layer ---
ax4 = fig.add_subplot(gs[1, 1])
proj0 = sdata["projections"][0]            # embedding layer
proj_best = sdata["projections"][bl + 1]   # after best encoder layer (hidden_states offset by 1)

# Layer 0 points (gray, smaller)
ax4.scatter(proj0[:, 0], proj0[:, 1], c="#CCCCCC", s=30, zorder=1, label=f"Layer 0 (embedding)")
for i, tok in enumerate(tokens):
    if sdata["special_token_mask"][i]:
        continue
    ax4.annotate(tok, (proj0[i, 0], proj0[i, 1]),
                 fontsize=5, color="#999999", xytext=(3, 3), textcoords="offset points")

# Best-layer points (colored)
for i, tok in enumerate(tokens):
    if sdata["special_token_mask"][i]:
        continue
    is_bank = tok == "bank"
    color = "#E74C3C" if is_bank else "#2980B9"
    size = 60 if is_bank else 25
    ax4.scatter(proj_best[i, 0], proj_best[i, 1], c=color, s=size, zorder=2, edgecolors="white")
    ax4.annotate(tok, (proj_best[i, 0], proj_best[i, 1]),
                 fontsize=6, fontweight="bold" if is_bank else "normal",
                 color=color, xytext=(3, 3), textcoords="offset points")

    # Draw movement arrow from layer 0 to best layer
    if not sdata["special_token_mask"][i]:
        ax4.annotate("", xy=(proj_best[i, 0], proj_best[i, 1]),
                     xytext=(proj0[i, 0], proj0[i, 1]),
                     arrowprops=dict(arrowstyle="->", color="#AAAAAA", lw=0.5, alpha=0.4))

ax4.set_title(f"Token positions: embedding → layer {bl}", fontsize=10)
ax4.legend(fontsize=7)
ax4.set_aspect("equal")

fig.suptitle(f"Attention Data Validation — \"{sdata['text'][:60]}...\"", fontsize=12, y=1.01)
plt.tight_layout()
plt.show()

# --- Print summary ---
print("\n" + "=" * 60)
print("VALIDATION SUMMARY")
print("=" * 60)

for sid, sdata in SENTENCES.items():
    bl = best_defaults[sid]["layer"]
    bh = best_defaults[sid]["head"]
    print(f"\n{sid} (best: layer {bl}, head {bh}):")
    for toi in sdata["tokens_of_interest"]:
        ctx_nn = toi["contextual_neighbors"][bl][:5]
        static_nn = toi["static_neighbors"][:5]
        ctx_words = [nb["word"] for nb in ctx_nn]
        static_words = [nb["word"] for nb in static_nn]
        print(f'  {toi["label"]}:')
        print(f"    Static neighbors:      {static_words}")
        print(f"    Contextual (layer {bl:2d}): {ctx_words}")
````

````python
# ============================================================
# 13. EXPORT JSON
# ============================================================

def round_list(arr, decimals):
    """Convert numpy array to nested Python lists with rounding."""
    if isinstance(arr, np.ndarray):
        return np.round(arr, decimals).tolist()
    return arr


print("Exporting attention_data.json ...\n")

export = {
    "metadata": {
        "model": MODEL_NAME,
        "nLayers": n_layers,
        "nHeads": n_heads,
        "headDim": head_dim,
        "hiddenDim": hidden_dim,
        "qkvExportLayers": qkv_layers,
        "nNeighbors": N_NEIGHBORS,
    },
    "bestDefaults": best_defaults,
    "sentences": {},
}

for sid, sdata in SENTENCES.items():
    seq_len = len(sdata["tokens"])

    # --- Attention weights: all layers, all heads ---
    # Shape: [n_layers][n_heads][seq_len][seq_len]
    attn_export = []
    for li in range(n_layers):
        heads = []
        for hi in range(n_heads):
            heads.append(round_list(sdata["attentions"][li][hi], ATTN_PRECISION))
        attn_export.append(heads)

    # --- Raw QK scores: all layers, all heads ---
    # Shape: [n_layers][n_heads][seq_len][seq_len]
    scores_export = []
    for li in range(n_layers):
        heads = []
        for hi in range(n_heads):
            heads.append(round_list(sdata["qk_scores_all"][li][hi], SCORE_PRECISION))
        scores_export.append(heads)

    # --- Q/K/V vectors: selected layers only ---
    # Shape per layer: {q, k, v} each [n_heads][seq_len][head_dim]
    qkv_export = {}
    for li in qkv_layers:
        qkv_data = sdata["qkv_all"][li]
        qkv_export[str(li)] = {
            "q": round_list(qkv_data["q"], QKV_PRECISION),  # [n_heads, seq_len, head_dim]
            "k": round_list(qkv_data["k"], QKV_PRECISION),
            "v": round_list(qkv_data["v"], QKV_PRECISION),
        }

    # --- Tokens of interest with neighbors ---
    toi_export = []
    for toi in sdata["tokens_of_interest"]:
        ctx_neighbors = {}
        for layer_idx in range(n_layers + 1):
            ctx_neighbors[str(layer_idx)] = toi["contextual_neighbors"][layer_idx]

        toi_export.append({
            "index": toi["index"],
            "token": toi["token"],
            "label": toi["label"],
            "contextType": toi["context_type"],
            "bestLayer": toi.get("best_layer", best_defaults[sid]["layer"]),
            "bestHead": toi.get("best_head", best_defaults[sid]["head"]),
            "staticNeighbors": toi["static_neighbors"],
            "contextualNeighbors": ctx_neighbors,
        })

    # --- 2D sentence projections ---
    proj_export = {}
    for layer_idx in range(n_layers + 1):
        proj_export[str(layer_idx)] = round_list(
            sdata["projections"][layer_idx], PROJ_PRECISION
        )

    # --- Assemble sentence export ---
    export["sentences"][sid] = {
        "text": sdata["text"],
        "tokens": sdata["tokens"],
        "tokenIds": sdata["token_ids"],
        "specialTokenMask": sdata["special_token_mask"],
        "attentionWeights": attn_export,
        "qkScores": scores_export,
        "qkv": qkv_export,
        "tokensOfInterest": toi_export,
        "sentenceProjections": proj_export,
    }

    print(f"  ✓ {sid}: {seq_len} tokens, "
          f"{len(attn_export)} layers × {len(attn_export[0])} heads attention, "
          f"QKV for layers {qkv_layers}, "
          f"{len(toi_export)} tokens of interest")

# --- Write to file ---
out_path = OUTPUT_DIR / "attention_data.json"
with open(out_path, "w") as f:
    json.dump(export, f, separators=(",", ":"))

file_size_mb = out_path.stat().st_size / 1024 / 1024
print(f"\n✓ {out_path}  ({file_size_mb:.1f} MB)")

# --- Size breakdown ---
print("\n--- Approximate size breakdown ---")
for sid in export["sentences"]:
    sent_json = json.dumps(export["sentences"][sid], separators=(",", ":"))
    print(f"  {sid}: {len(sent_json) / 1024:.0f} KB")
````

````python
# ============================================================
# 14. CLEANUP — free large in-memory objects
# ============================================================
# The raw numpy arrays are no longer needed once the JSON is written.

for sdata in SENTENCES.values():
    del sdata["hidden_states"]
    del sdata["attentions"]
    del sdata["qkv_all"]
    del sdata["qk_scores_all"]
    del sdata["projections"]
    del sdata["encoded"]

del word_embeddings, ref_embeddings, ref_embeddings_normed
print("✓ In-memory data cleaned up.")

print(f"\n{'=' * 60}")
print(f"OUTPUT FILES")
print(f"{'=' * 60}")
print(f"  {OUTPUT_DIR.resolve()}/attention_data.json  ({file_size_mb:.1f} MB)")
print(f"\n  Ready for the Session 2 HTML/JS demo!")
````
