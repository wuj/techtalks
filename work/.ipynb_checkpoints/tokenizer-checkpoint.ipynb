````python
# ============================================================
# 2. CONFIGURATION
# ============================================================
import json
from pathlib import Path

# --- Primary tokenizer (used in the JS demo) ---
# cl100k_base = GPT-4 / GPT-3.5-turbo / text-embedding-ada-002
# This is what js-tiktoken / gpt-tokenizer implements client-side.
PRIMARY_ENCODING = "cl100k_base"
PRIMARY_LABEL = "GPT-4"

# --- Comparison tokenizers ---
COMPARE_ENCODINGS = {
    "gpt2": {
        "type": "tiktoken",
        "encoding": "r50k_base",   # original GPT-2
        "label": "GPT-2",
    },
    "bert": {
        "type": "transformers",
        "model": "bert-base-uncased",
        "label": "BERT (WordPiece)",
    },
}

# --- Output ---
OUTPUT_DIR = Path("/home/jovyan/demo_data")
OUTPUT_DIR.mkdir(exist_ok=True)
````

````python
# ============================================================
# 3. LOAD TOKENIZERS
# ============================================================
import tiktoken
from transformers import BertTokenizer

# --- Primary: GPT-4 ---
primary_enc = tiktoken.get_encoding(PRIMARY_ENCODING)

# --- Comparison: GPT-2 ---
gpt2_enc = tiktoken.get_encoding(COMPARE_ENCODINGS["gpt2"]["encoding"])

# --- Comparison: BERT ---
bert_tokenizer = BertTokenizer.from_pretrained(
    COMPARE_ENCODINGS["bert"]["model"]
)

print(f"âœ“ Primary: {PRIMARY_LABEL} ({PRIMARY_ENCODING})")
print(f"  Vocab size: {primary_enc.n_vocab:,}")
print(f"âœ“ GPT-2 ({COMPARE_ENCODINGS['gpt2']['encoding']})")
print(f"  Vocab size: {gpt2_enc.n_vocab:,}")
print(f"âœ“ BERT ({COMPARE_ENCODINGS['bert']['model']})")
print(f"  Vocab size: {bert_tokenizer.vocab_size:,}")
````

````python
# ============================================================
# 4. TOKENIZER HELPER FUNCTIONS
# ============================================================

def tokenize_primary(text: str) -> list[dict]:
    """Tokenize with the primary encoding, returning rich per-token data."""
    token_ids = primary_enc.encode(text, allowed_special="all")
    tokens = []
    byte_offset = 0

    for tid in token_ids:
        token_bytes = primary_enc.decode_single_token_bytes(tid)
        token_str = token_bytes.decode("utf-8", errors="replace")
        tokens.append({
            "id": tid,
            "text": token_str,
            "displayText": repr(token_str) if not token_str.isprintable()
                           or token_str.isspace() and token_str != " "
                           else token_str,
            "byteLen": len(token_bytes),
        })
        byte_offset += len(token_bytes)

    return tokens


def tokenize_gpt2(text: str) -> list[dict]:
    """Tokenize with GPT-2 encoding."""
    token_ids = gpt2_enc.encode(text)
    tokens = []
    for tid in token_ids:
        token_bytes = gpt2_enc.decode_single_token_bytes(tid)
        token_str = token_bytes.decode("utf-8", errors="replace")
        tokens.append({"id": tid, "text": token_str})
    return tokens


def tokenize_bert(text: str) -> list[dict]:
    """Tokenize with BERT WordPiece (no [CLS]/[SEP])."""
    token_strs = bert_tokenizer.tokenize(text)
    token_ids = bert_tokenizer.convert_tokens_to_ids(token_strs)
    return [{"id": tid, "text": ts} for ts, tid in zip(token_strs, token_ids)]


# Quick validation
test = "Hello, world!"
print(f'\nTest: "{test}"')
print(f"  {PRIMARY_LABEL}: {[(t['text'], t['id']) for t in tokenize_primary(test)]}")
print(f"  GPT-2:  {[(t['text'], t['id']) for t in tokenize_gpt2(test)]}")
print(f"  BERT:   {[(t['text'], t['id']) for t in tokenize_bert(test)]}")
````

````python
# ============================================================
# 6. TOKENIZE ALL CURATED EXAMPLES
# ============================================================

# Category display metadata
CATEGORY_META = {
    "subword_splitting": {
        "label": "Subword Splitting",
        "description": "How words break into pieces",
        "icon": "âœ‚ï¸",
        "order": 0,
    },
    "names": {
        "label": "Names",
        "description": "Personal and proper names",
        "icon": "ğŸ‘¤",
        "order": 1,
    },
    "code": {
        "label": "Code Snippets",
        "description": "Programming languages and markup",
        "icon": "ğŸ’»",
        "order": 2,
    },
    "slang": {
        "label": "Slang & Internet",
        "description": "Informal language and internet culture",
        "icon": "ğŸ—£ï¸",
        "order": 3,
    },
    "emoji_unicode": {
        "label": "Emoji & Unicode",
        "description": "Emoji, special characters, multi-byte sequences",
        "icon": "ğŸ¨",
        "order": 4,
    },
    "multilingual": {
        "label": "Multilingual",
        "description": "Non-English scripts and mixed languages",
        "icon": "ğŸŒ",
        "order": 5,
    },
    "numbers": {
        "label": "Numbers & Math",
        "description": "Digits, arithmetic, numerical patterns",
        "icon": "ğŸ”¢",
        "order": 6,
    },
    "whitespace": {
        "label": "Whitespace & Formatting",
        "description": "Spaces, newlines, indentation",
        "icon": "â¬œ",
        "order": 7,
    },
    "edge_cases": {
        "label": "Edge Cases & Fun",
        "description": "Unusual inputs and linguistic puzzles",
        "icon": "ğŸ§©",
        "order": 8,
    },
}

for example in CURATED_EXAMPLES:
    text = example["text"]

    # --- Primary tokenizer ---
    tokens = tokenize_primary(text)
    example["primaryTokens"] = tokens
    example["primaryTokenCount"] = len(tokens)
    example["charCount"] = len(text)
    example["compressionRatio"] = round(len(text) / max(len(tokens), 1), 2)

    # --- Comparison tokenizers ---
    gpt2_tokens = tokenize_gpt2(text)
    bert_tokens = tokenize_bert(text) if text.strip() else []
    example["comparison"] = {
        "gpt2": {
            "tokens": gpt2_tokens,
            "tokenCount": len(gpt2_tokens),
        },
        "bert": {
            "tokens": bert_tokens,
            "tokenCount": len(bert_tokens),
        },
    }

print("Tokenized all examples. Sample:\n")

# Show a few interesting ones
for ex in CURATED_EXAMPLES[:3]:
    text = ex["text"]
    ptok = ex["primaryTokens"]
    display = " | ".join(t["text"] for t in ptok)
    ids = [t["id"] for t in ptok]
    print(f'  "{text}"')
    print(f"    {PRIMARY_LABEL} ({len(ptok)} tokens): [{display}]")
    print(f"    IDs: {ids}")
    print(f"    GPT-2 ({ex['comparison']['gpt2']['tokenCount']} tokens): "
          f"[{' | '.join(t['text'] for t in ex['comparison']['gpt2']['tokens'])}]")
    print(f"    BERT  ({ex['comparison']['bert']['tokenCount']} tokens): "
          f"[{' | '.join(t['text'] for t in ex['comparison']['bert']['tokens'])}]")
    print()
````

````python
# ============================================================
# 7. ONE-HOT ENCODING DEMO DATA
# ============================================================
# The Session 1 plan shows one-hot vectors for "king" and "queen"
# side by side, walks through the dot product, then adds
# "refrigerator" to show kingÂ·queen = kingÂ·refrigerator = 0.
#
# We prepare:
#   - The actual token IDs (positions of the "1" in the one-hot vector)
#   - A simplified vocab-size for visualization (we use the real vocab
#     size but suggest the JS demo shows a "zoomed" window of ~20 slots
#     centered on the relevant IDs)
#   - A pre-built dot product animation sequence

one_hot_words = ["king", "queen", "refrigerator", "man", "woman"]

one_hot_data = {
    "vocabSize": primary_enc.n_vocab,
    "words": {},
    "dotProducts": [],
    "animationSequence": [],
}

for word in one_hot_words:
    tokens = primary_enc.encode(word)
    # If the word is a single token, we can do one-hot cleanly
    is_single_token = len(tokens) == 1
    one_hot_data["words"][word] = {
        "tokenIds": tokens,
        "isSingleToken": is_single_token,
        "primaryTokenId": tokens[0],
        "tokenTexts": [primary_enc.decode([tid]) for tid in tokens],
    }
    status = "âœ“ single token" if is_single_token else f"âš  {len(tokens)} tokens"
    print(f'  "{word}": ID={tokens[0]:<8d} {status}')

# Pre-compute all pairwise dot products
# (They're all 0 for different tokens, but we make it explicit for the animation)
print("\nDot products:")
pairs = [
    ("king", "queen",        "Semantically similar â€” but dot product is 0!"),
    ("king", "refrigerator", "Semantically unrelated â€” also 0. Identical to kingÂ·queen."),
    ("king", "king",         "Same word â€” dot product is 1. The only non-zero case."),
    ("man",  "woman",        "Another related pair â€” still 0."),
]

for word_a, word_b, explanation in pairs:
    id_a = one_hot_data["words"][word_a]["primaryTokenId"]
    id_b = one_hot_data["words"][word_b]["primaryTokenId"]
    dot_product = 1 if id_a == id_b else 0

    entry = {
        "wordA": word_a,
        "wordB": word_b,
        "tokenIdA": id_a,
        "tokenIdB": id_b,
        "dotProduct": dot_product,
        "explanation": explanation,
    }
    one_hot_data["dotProducts"].append(entry)
    print(f"  {word_a:15s} Â· {word_b:15s} = {dot_product}  ({explanation})")

# Animation sequence: step-by-step instructions for the JS demo
one_hot_data["animationSequence"] = [
    {
        "step": 1,
        "action": "show_vector",
        "word": "king",
        "narration": "Here's the one-hot vector for 'king'. It has a 1 at one position "
                     "and 0s everywhere else.",
    },
    {
        "step": 2,
        "action": "show_vector",
        "word": "queen",
        "narration": "Now 'queen'. Its 1 is at a different position.",
    },
    {
        "step": 3,
        "action": "compute_dot_product",
        "wordA": "king",
        "wordB": "queen",
        "narration": "Dot product: multiply element-by-element, then sum. "
                     "The 1s are in different positions, so every product is 0. "
                     "Total: 0. To this representation, 'king' and 'queen' are "
                     "completely unrelated.",
    },
    {
        "step": 4,
        "action": "show_vector",
        "word": "refrigerator",
        "narration": "Now 'refrigerator' â€” obviously unrelated to 'king.'",
    },
    {
        "step": 5,
        "action": "compute_dot_product",
        "wordA": "king",
        "wordB": "refrigerator",
        "narration": "Dot product with 'king': also 0. Exactly the same as kingÂ·queen. "
                     "This representation has no concept of meaning. That's the problem.",
    },
    {
        "step": 6,
        "action": "reveal_problem",
        "narration": "Every pair of different words has similarity 0. "
                     "There's no structure, no notion of closeness. "
                     "We need something better: embeddings.",
    },
]

print(f"\nâœ“ One-hot data prepared for {len(one_hot_words)} words, "
      f"{len(pairs)} dot products, {len(one_hot_data['animationSequence'])} animation steps")
````

````python
# ============================================================
# 8. VOCABULARY STATISTICS
# ============================================================
# Useful context for the presenter and for potential UI display.

import statistics

# Compute stats across all curated examples
primary_counts = [e["primaryTokenCount"] for e in CURATED_EXAMPLES if e["text"]]
gpt2_counts = [e["comparison"]["gpt2"]["tokenCount"] for e in CURATED_EXAMPLES if e["text"]]
bert_counts = [e["comparison"]["bert"]["tokenCount"] for e in CURATED_EXAMPLES if e["text"]]

vocab_stats = {
    "tokenizers": {
        PRIMARY_LABEL: {
            "encoding": PRIMARY_ENCODING,
            "vocabSize": primary_enc.n_vocab,
            "type": "BPE (byte-level)",
        },
        "GPT-2": {
            "encoding": COMPARE_ENCODINGS["gpt2"]["encoding"],
            "vocabSize": gpt2_enc.n_vocab,
            "type": "BPE (byte-level)",
        },
        "BERT (WordPiece)": {
            "model": COMPARE_ENCODINGS["bert"]["model"],
            "vocabSize": bert_tokenizer.vocab_size,
            "type": "WordPiece",
        },
    },
    "exampleStats": {
        "totalExamples": len(CURATED_EXAMPLES),
        "avgTokensPrimary": round(statistics.mean(primary_counts), 1),
        "avgTokensGpt2": round(statistics.mean(gpt2_counts), 1),
        "avgTokensBert": round(statistics.mean(bert_counts), 1),
        "maxTokensPrimary": max(primary_counts),
        "minTokensPrimary": min(primary_counts),
    },
}

print("Vocabulary statistics:")
for name, info in vocab_stats["tokenizers"].items():
    print(f"  {name}: {info.get('vocabSize', 'N/A'):,} tokens ({info['type']})")

print(f"\nExample stats (non-empty):")
print(f"  Avg tokens ({PRIMARY_LABEL}): {vocab_stats['exampleStats']['avgTokensPrimary']}")
print(f"  Avg tokens (GPT-2): {vocab_stats['exampleStats']['avgTokensGpt2']}")
print(f"  Avg tokens (BERT):  {vocab_stats['exampleStats']['avgTokensBert']}")
````

````python
# ============================================================
# 9. CROSS-TOKENIZER COMPARISON HIGHLIGHTS
# ============================================================
# Find examples where tokenizers disagree most â€” these make the
# best presentation talking points.

print("Cross-tokenizer comparison (biggest disagreements):\n")

comparisons = []
for ex in CURATED_EXAMPLES:
    if not ex["text"]:
        continue
    p = ex["primaryTokenCount"]
    g = ex["comparison"]["gpt2"]["tokenCount"]
    b = ex["comparison"]["bert"]["tokenCount"]
    spread = max(p, g, b) - min(p, g, b)
    comparisons.append({
        "text": ex["text"][:50] + ("..." if len(ex["text"]) > 50 else ""),
        "primary": p,
        "gpt2": g,
        "bert": b,
        "spread": spread,
    })

# Sort by spread descending
comparisons.sort(key=lambda x: -x["spread"])

print(f"  {'Text':<53s} {PRIMARY_LABEL:>5s} {'GPT2':>5s} {'BERT':>5s} {'Î”':>3s}")
print(f"  {'â”€' * 53} {'â”€' * 5} {'â”€' * 5} {'â”€' * 5} {'â”€' * 3}")
for c in comparisons[:12]:
    print(f"  {c['text']:<53s} {c['primary']:>5d} {c['gpt2']:>5d} "
          f"{c['bert']:>5d} {c['spread']:>3d}")
````

````python
# ============================================================
# 12. VALIDATION â€” verify JS tokenizer will match
# ============================================================
# The critical invariant: the JS demo uses the same cl100k_base
# encoding. If someone types "unbelievably" live, the result must
# match what we've pre-computed here. Let's verify round-trip
# consistency.

print("Validation: round-trip consistency check\n")

all_match = True
for ex in CURATED_EXAMPLES:
    text = ex["text"]
    expected_ids = [t["id"] for t in ex["primaryTokens"]]

    # Re-encode from scratch
    actual_ids = primary_enc.encode(text, allowed_special="all")

    # Decode back to text
    reconstructed = primary_enc.decode(actual_ids)

    ids_match = expected_ids == actual_ids
    text_match = reconstructed == text

    if not ids_match or not text_match:
        print(f'  âš  MISMATCH: "{text[:40]}..."')
        if not ids_match:
            print(f"    Expected IDs: {expected_ids[:5]}...")
            print(f"    Actual IDs:   {actual_ids[:5]}...")
        if not text_match:
            print(f"    Original:      {repr(text[:40])}")
            print(f"    Reconstructed: {repr(reconstructed[:40])}")
        all_match = False

if all_match:
    print(f"  âœ“ All {len(CURATED_EXAMPLES)} examples pass round-trip validation.")
    print(f"    The JS tokenizer (cl100k_base) will produce identical results.")
else:
    print(f"\n  âš  Some mismatches found. Check the examples above.")

# --- Show the "opener" example one more time ---
opener = next(e for e in CURATED_EXAMPLES if e.get("isOpener"))
print(f'\n--- Opener example for presentation ---')
print(f'  "{opener["text"]}"')
tok_display = " | ".join(
    f'{t["text"]} [{t["id"]}]' for t in opener["primaryTokens"]
)
print(f"  â†’ {tok_display}")
print(f"  Annotation: {opener['annotation']}")

print(f"\n{'=' * 60}")
print(f"OUTPUT FILE")
print(f"{'=' * 60}")
print(f"  {OUTPUT_DIR.resolve()}/tokenizer_examples.json  ({file_size_kb:.0f} KB)")
print(f"\n  Ready for the Session 1 HTML/JS demo!")
````
